```python
f=open("projet.csv")
file=csv.reader(f)
data=list(file)
data
header=data[0]
data=data[1:]
header
data[:5]
dic_kills={}
years=[row[1] for row in data]
for year in years:
    if year in dic_kills:
        dic_kills[year]+=1
    else:
        dic_kills[year]=1
dic_kills
```

```python
import datetime
dates=[datetime.datetime(year=int(row[1]),month=int(row[2]),day=1) for row in data]
dates
```

```python
dic_dates={}
for date in dates:
    if(date not in dic_dates):
        dic_dates[date]=0
    else:
        dic_dates[date]+=1
dic_dates
```

```python
sexe_count={}
sexes=[row[5] for row in data]
for sex in sexes:
    if sex not in sexe_count:
        sexe_count[sex]=1
    else: sexe_count[sex]+=1

races_count={}
for race in races:
    if race not in races_count:
        races_count[race]=1
    else: races_count[race]+=1
        
races_count
```
