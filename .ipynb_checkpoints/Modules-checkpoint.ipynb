```python
import csv
f=open("nfl_csv.csv")
obj= csv.reader(f)
mylist=list(obj)

def victory_per_year(team,year):
    number_of_vic=0
    for line in mylist:
        if line[0]==year and line[2]==team:
            number_of_vic+=1
    return number_of_vic

print(victory_per_year("Chicago Bears","2009"))
```
